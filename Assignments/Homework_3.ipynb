R3.23 Give a big-Oh characterization, in terms of n, of the running time of the
example2 function shown in Code Fragment 3.10

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def example(S):
    n = len(S)
    total = 0
    count = 0
    numbers = []
    for j in range(n):
        total += S[j]
        count += 1
        numbers.append(count)
    return total, numbers
```

```python
s = np.random.randint(0, 100, size=(100,))
```

```python
total, numbers = example(s)
plt.plot(numbers)
```

R3.27 Give a big-Oh characterization, in terms of n, of the running time of the example5 function.

```python
def example5(A, B):
    """Return the number of elements in B equal to the sum of prefix sums in A."""
    n = len(A)
    count = 0
    num_calc = 0
    for i in range(n):
        total = 0
        for j in range(n):
            for k in range(1+j):
                total += A[k]
                num_calc += 1
        if B[i] == total:
            count+=1
    return count, num_calc
def example5_a(n=100):
    op_nums = []
    for i in range(1, n):
        A = np.random.randint(0, 100, size=(i, ))
        B = np.random.randint(0, 100, size=(i, ))
        _, num_calc = example5(A, B)
        op_nums.append(num_calc)
    plt.plot(op_nums)
    plt.show()
example5_a()
```
